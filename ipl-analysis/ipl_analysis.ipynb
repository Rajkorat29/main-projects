```python
# IPL Data Analysis Project (2008â€“2024)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
matches = pd.read_csv("matches.csv")

# Basic info
print("Total matches:", matches.shape[0])
print("Columns:", matches.columns)

# Preview the data
matches.head()
```

```python
team_wins = matches['winner'].value_counts().head(5)
print(team_wins)

plt.figure(figsize=(8, 5))
sns.barplot(x=team_wins.index, y=team_wins.values, palette='viridis')
plt.title("Top 5 Teams with Most Wins")
plt.ylabel("Wins")
plt.xlabel("Team")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
matches['toss_win_match_win'] = matches['toss_winner'] == matches['winner']
percent = matches['toss_win_match_win'].value_counts(normalize=True) * 100
print(percent)

labels = ['Toss & Match Win', 'Toss Win but Match Lose']
plt.pie(percent, labels=labels, autopct='%1.1f%%', startangle=140, colors=['#00cc99', '#ff6666'])
plt.title("Toss Winner also Match Winner?")
plt.show()




```

```python
top_players = matches['player_of_match'].value_counts().head(10)

plt.figure(figsize=(8, 5))
sns.barplot(x=top_players.values, y=top_players.index, palette='coolwarm')
plt.title("Top 10 Players with Most 'Player of the Match' Awards")
plt.xlabel("Awards")
plt.ylabel("Player")
plt.tight_layout()
plt.show()
```

```python
season_match_count = matches['season'].value_counts().sort_index()

plt.figure(figsize=(10, 5))
sns.lineplot(x=season_match_count.index, y=season_match_count.values, marker='o')
plt.title("Total Matches per IPL Season")
plt.xlabel("Season")
plt.ylabel("Matches")
plt.grid(True)
plt.tight_layout()
plt.show()
```
